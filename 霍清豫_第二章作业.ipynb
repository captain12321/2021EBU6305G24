```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#使用histogram画CRIM特征
sns.histplot(df["crim"],bins=20)
plt.xlabel("crime_rate")
```

```python
#使用histogram画ZN特征
sns.histplot(df["zn"],bins=20)
plt.xlabel("ZN")
```

```python
#使用histogram画INDUS特征
sns.histplot(df["indus"],bins=40)
plt.xlabel("INDUS")
```

```python
#使用barplot画CHAS特征
count_1=df["chas"].sum()
count_0=df.shape[0]-count_1
sns.barplot(x =[0,1], y =[count_0,count_1])
plt.xlabel("CHAS")
```

```python
#使用histogram画NOX,RM,AGE,DIS,RAD,TAX,MEDV特征
sns.histplot(df["nox"],bins=30)
plt.xlabel("nox")
```

```python
#用KDE画PTRATIO，LSTAT
sns.kdeplot(df["ptratio"],fill=True)
plt.xlabel=("ptratio")
```

```python
sns.kdeplot(df["lstat"],fill=True)
plt.xlabel=("lstat_percentage")
```

```python
#利用scatter plot 画出各个特征和medv之间的关系
for column in df.columns:
    if column != 'medv':
        sns.scatterplot(x=df[column], y=df['medv'])
        plt.title(f'Relationship between {column} and medv')
        plt.show()
```

```python
#由图2可以看出房价的中位数（medv）和犯罪率（crime）成负相关
#由图4可以看出房价中位数和INDUS间成略微负相关
#由图6可以看出房价中位数和一氧化氮排放量（nox）成负相关
#由图7可以看出房价中位数和房间数量（rm）成正相关
#由最后一张图可以看出房价中位数和低收入人群比例（lstat）成负相关
```

```python
#分析各个特征间的关系
#1.dis和nox之间的关系图
sns.scatterplot(x=df["dis"], y=df['nox'])
#看出nox和dis间成负相关关系，离公路越远一氧化氮含量越低
```

```python
#2.age和nox的关系

#nox和age之间成略微正相关，房龄越老，一氧化氮含量越高
```

```python
#3.dis和indus
sns.scatterplot(x=df["dis"], y=df['indus'])
#indus和dis间的关系成负相关，离公路越远非零售商业用地越少
```

```python
#4.indus和zn
sns.scatterplot(x=df["zn"], y=df['indus'])
#看出indus和zn成负相关，住房用地越大，非零售商业用地越少
```

```python
#5.dis和crim
sns.scatterplot(x=df["dis"], y=df['crim'])
#crim和dis成负相关，看出距离公路越远犯罪率越低
```

```python
#利用hexplot画mdev和lstat间的关系
sns.jointplot(x=df["lstat"],y=df["medv"],kind="hex")
```

```python
#利用Multivariate Scatter Plot画chas medv和rm之间的关系
sns.scatterplot(x=df["rm"],y=df["medv"],hue=df["chas"])
#可以看出随着房间数的增长房价中位数上升，且我们发现有路能到Charles河的房子房间数都大于5间
```
